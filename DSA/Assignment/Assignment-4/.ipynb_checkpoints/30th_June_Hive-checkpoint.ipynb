1.  Write a Python program that uses the HiveQL language to create a table named "Employees" with columns for "id," "name," and "salary."

```python
from pyhive import hive

# Establish a connection to Hive
conn = hive.Connection(host='localhost', port=10000, username='your_username')

# Create a cursor object
cursor = conn.cursor()

# HiveQL query to create the Employees table
create_table_query = """
    CREATE TABLE Employees (
        id INT,
        name STRING,
        salary FLOAT
    )
"""

# Execute the query
cursor.execute(create_table_query)

# Commit the changes
conn.commit()

# Close the cursor and connection
cursor.close()
conn.close()
```

2.  Create a Python program that retrieves records from a Hive table named "Customers" where the age is greater than 30.

```python
from pyhive import hive

# Establish a connection to Hive
conn = hive.Connection(host='localhost', port=10000, username='your_username')

# Create a cursor object
cursor = conn.cursor()

# HiveQL query to retrieve records from the Customers table where age is greater than 30
retrieve_records_query = """
    SELECT *
    FROM Customers
    WHERE age > 30
"""

# Execute the query
cursor.execute(retrieve_records_query)

# Fetch all the rows returned by the query
rows = cursor.fetchall()

# Process the rows
for row in rows:
    # Access the columns by index or name
    id = row[0]
    name = row[1]
    age = row[2]
    
    # Do something with the retrieved data
    print(f"ID: {id}, Name: {name}, Age: {age}")

# Close the cursor and connection
cursor.close()
conn.close()
```

3.  Write a Python script that sorts records in descending order based on the "timestamp" column in a Hive table named "Logs."


```python
from pyhive import hive

# Establish a connection to Hive
conn = hive.Connection(host='localhost', port=10000, username='your_username')

# Create a cursor object
cursor = conn.cursor()

# HiveQL query to sort records in descending order based on the "timestamp" column
sort_records_query = """
    SELECT *
    FROM Logs
    ORDER BY timestamp DESC
"""

# Execute the query
cursor.execute(sort_records_query)

# Fetch all the rows returned by the query
rows = cursor.fetchall()

# Process the rows
for row in rows:
    # Access the columns by index or name
    id = row[0]
    timestamp = row[1]
    message = row[2]
    
    # Do something with the retrieved data
    print(f"ID: {id}, Timestamp: {timestamp}, Message: {message}")

# Close the cursor and connection
cursor.close()
conn.close()
```

5.  Write a Python script that calculates the average salary of employees from a Hive table named "Employees".


```python
#pip install pyhive

from pyhive import hive

# Establish a connection to Hive
conn = hive.Connection(host='localhost', port=10000, username='your_username')

# Create a cursor object
cursor = conn.cursor()

# HiveQL query to calculate the average salary of employees
average_salary_query = """
    SELECT AVG(salary) as average_salary
    FROM Employees
"""

# Execute the query
cursor.execute(average_salary_query)

# Fetch the result returned by the query
result = cursor.fetchone()

# Access the average salary from the result
average_salary = result[0]

# Print the average salary
print(f"Average Salary: {average_salary}")

# Close the cursor and connection
cursor.close()
conn.close()

```

6.  Implement a Python program that uses Hive partitioning to create a partitioned table named "Sales_Data" based on the "year" and "month" columns.



```python
from pyhive import hive

# Establish a connection to Hive
conn = hive.Connection(host='localhost', port=10000, username='your_username')

# Create a cursor object
cursor = conn.cursor()

# HiveQL query to create the partitioned table
create_table_query = """
    CREATE TABLE Sales_Data (
        id INT,
        product STRING,
        price FLOAT
    )
    PARTITIONED BY (year INT, month INT)
"""

# Execute the query to create the table
cursor.execute(create_table_query)

# HiveQL query to add partitions to the table
add_partitions_query = """
    ALTER TABLE Sales_Data
    ADD PARTITION (year=2022, month=1)
    ADD PARTITION (year=2022, month=2)
    ADD PARTITION (year=2022, month=3)
"""

# Execute the query to add partitions
cursor.execute(add_partitions_query)

# Commit the changes
conn.commit()

# Close the cursor and connection
cursor.close()
conn.close()
```

8.  Create a Python program that performs an inner join between two Hive tables, "Orders" and "Customers," based on a common column.



```python
from pyhive import hive

# Establish a connection to Hive
conn = hive.Connection(host='localhost', port=10000, username='your_username')

# Create a cursor object
cursor = conn.cursor()

# HiveQL query to perform the inner join between Orders and Customers tables
inner_join_query = """
    SELECT *
    FROM Orders
    INNER JOIN Customers ON Orders.customer_id = Customers.customer_id
"""

# Execute the query
cursor.execute(inner_join_query)

# Fetch all the rows returned by the query
rows = cursor.fetchall()

# Process the rows
for row in rows:
    # Access the columns by index or name
    order_id = row[0]
    customer_id = row[1]
    order_date = row[2]
    customer_name = row[3]
    customer_address = row[4]
    
    # Do something with the joined data
    print(f"Order ID: {order_id}, Customer ID: {customer_id}, Order Date: {order_date}, Customer Name: {customer_name}, Customer Address: {customer_address}")

# Close the cursor and connection
cursor.close()
conn.close()
```
