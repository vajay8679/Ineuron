There is a singly-linked listÂ `head`Â and we want to delete a nodeÂ `node`Â in it.

You are given the node to be deletedÂ `node`. You willÂ **not be given access**Â to the first node ofÂ `head`.

Delete the given node. Note that by deleting the node, we do not mean removing it from memory. We mean:

# **Approach:**

We first update the value of the current node with the value of its next node, and then we skip the next node by pointing the current node's next pointer to its next node's next pointer.

</aside>

# **Complexity:**

- Time complexity: Since we are only modifying the value and next pointer of the given node, the time complexity of this algorithm is O(1).
- Space complexity: As we are not using any extra space, the space complexity of this algorithm is O(1).

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def deleteNode(node):
    node.val = node.next.val
    node.next = node.next.next
````

````python
# Example 1
head = ListNode(4)
node = ListNode(5)
head.next = node
node.next = ListNode(1)
node.next.next = ListNode(9)
deleteNode(node)
# After deleting the node with value 5
# The linked list should become 4 -> 1 -> 9
current = head
while current is not None:
    print(current.val, end=" ")
    current = current.next
# Output: 4 1 9

# Example 2
head = ListNode(4)
head.next = ListNode(5)
node = head.next
head.next.next = ListNode(1)
head.next.next.next = ListNode(9)
deleteNode(node)
# After deleting the node with value 1
# The linked list should become 4 -> 5 -> 9
current = head
while current is not None:
    print(current.val, end=" ")
    current = current.next
# Output: 4 5 9
````

ðŸ’¡ **Question 2**

Given theÂ `head`Â of a linked list and an integerÂ `val`, remove all the nodes of the linked list that hasÂ `Node.val == val`, and returnÂ *the new head*.

</aside>

Input: head = [1,2,6,3,4,5,6], val = 6
Output: [1,2,3,4,5]
    
Input: head = [], val = 1
Output: []

Input: head = [7,7,7,7], val = 7
Output: []

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeElements(head, val):
    # Handle empty list case
    if head is None:
        return None
    
    # Handle cases where the first few nodes have the given value
    while head is not None and head.val == val:
        head = head.next
    
    # Traverse the list and remove nodes with the given value
    current = head
    prev = None
    while current is not None:
        if current.val == val:
            prev.next = current.next
        else:
            prev = current
        current = current.next
    
    return head
````

````python
# Example 1
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(6)
head.next.next.next = ListNode(3)
head.next.next.next.next = ListNode(4)
head.next.next.next.next.next = ListNode(5)
head.next.next.next.next.next.next = ListNode(6)
val = 6
new_head = removeElements(head, val)
# After removing nodes with value 6
# The linked list should become 1 -> 2 -> 3 -> 4 -> 5
current = new_head
while current is not None:
    print(current.val, end=" ")
    current = current.next
# Output: 1 2 3 4 5

# Example 2
head = None
val = 1
new_head = removeElements(head, val)
# Since the list is empty, the result should be None
print(new_head)
# Output: None

# Example 3
head = ListNode(7)
head.next = ListNode(7)
head.next.next = ListNode(7)
head.next.next.next = ListNode(7)
val = 7
new_head = removeElements(head, val)
# Since all nodes have value 7, they will be removed
# The result should be None
print(new_head)
# Output: None
````

ðŸ’¡ **Question 3**

Given theÂ `head`Â of a singly linked list, reverse the list, and returnÂ *the reversed list*.

</aside>

Input: head = [1,2,3,4,5]
Output: [5,4,3,2,1]
    
Input: head = [1,2]
Output: [2,1]

Input: head = []
Output: []

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseList(head):
    prev = None
    current = head
    while current is not None:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    return prev
````

````python
# Example 1
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
reversed_head = reverseList(head)
# After reversing the list, it should become 5 -> 4 -> 3 -> 2 -> 1
current = reversed_head
while current is not None:
    print(current.val, end=" ")
    current = current.next
# Output: 5 4 3 2 1

# Example 2
head = ListNode(1)
head.next = ListNode(2)
reversed_head = reverseList(head)
# After reversing the list, it should become 2 -> 1
current = reversed_head
while current is not None:
    print(current.val, end=" ")
    current = current.next
# Output: 2 1

# Example 3
head = None
reversed_head = reverseList(head)
# Since the list is empty, the result should be None
print(reversed_head)
# Output: None
````
