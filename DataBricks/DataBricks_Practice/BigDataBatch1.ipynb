```
data = [(1,"Ajay","Male"),(2,"Amit","Male"),(3,"Sumit","Male"),(4,"Priya","Female"),(5,"Kusum","Female"),(6,"jaya","Female")]
col = ["id","name","gender"]

dfs = spark.createDataFrame(data,col)
display(dfs)
```

```
dfs.createOrReplaceTempView("person")
```

```
%sql
REMOVE WIDGET gender;
```

```
%sql
CREATE WIDGET DROPDOWN name DEFAULT 'Ajay' CHOICES SELECT  name FROM person; 
```

```
%sql
CREATE WIDGET DROPDOWN gender DEFAULT 'Male' CHOICES SELECT distinct gender FROM person; 
```

```
%sql
select * from person where gender = '$gender';
```
