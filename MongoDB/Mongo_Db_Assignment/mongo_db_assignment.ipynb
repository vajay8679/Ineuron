```python
import pandas as pd

dataframe=pd.read_csv("https://raw.githubusercontent.com/vigneshk/Admission-Dataset/master/Admission.csv")
```

```python
print(dataframe)
```

```python
import pymongo
```

```python
dbConn = pymongo.MongoClient("mongodb://localhost:27017/")
```

```python
db = dbConn["iNeuron"]
```

```python
collection = db['Addmision_details']
```

```python
dataframe.reset_index(inplace=True)
```

```python
data_to_dict = dataframe.to_dict("Records")
```

```python
#insert records
collection.insert_many(data_to_dict)
```

```python
all_records = collection.find()
```

```python
for i in all_records:
    print(i)
```
