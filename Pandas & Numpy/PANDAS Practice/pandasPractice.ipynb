# various way to create pandas library

```python
#create dict

d = {'name':['ajay','amit','sumit'],
     'age':[26,24,22],
      'addres':['indore','bhopal','chhindwara']}

# craete a dataframe (2d array - data strcture) for analysis big data easily
df1 = pd.DataFrame(d)
df1
```

```python
#create dict

d = {'name':['ajay','amit','sumit'],
     'age':[26,24,22],
      'addres':['indore','bhopal','chhindwara']}

# craete a dataframe (2d array - data strcture) for analysis big data easily
df1 = pd.DataFrame(d,index=['100','200','300'])
df1
```

```python
#create dict

d = {'name':['ajay','amit','sumit'],
     'age':[26,24,22],
      'addres':['indore','bhopal','chhindwara']}

# craete a dataframe (2d array - data strcture) for analysis big data easily
df1 = pd.DataFrame(d,index=pd.Index(['100','200','300'],name='INDEX'))
df1
```

```python
#create dict

d = {'name':['ajay','amit','sumit'],
     'age':[26,24,22],
      'address':['indore','bhopal','chhindwara']}
# craete a dataframe (2d array - data strcture) for analysis big data easily
df = pd.DataFrame(d,index=pd.Index(['100','200','300'],name='INDEX'),columns=pd.Index(['name','age','address'],name='COL'))
df
```
