```python
Binary Tree:-
* As the name says its a non-linear DS 
* Its a hierarchial DS
* Dynamic DS -> no limit to size
* A tree where each node has upto 2 children is a BT
    any node can have either {0, 1, 2} children
    0 children
    1 child -> left or right node
    2 children -> left node and right node
```

```python
Uses:-
* Flipkart -> Products (internally stored in a tree form)
    Apple -> ipad, mac, apple watch, ...
    Tables -> Coffee table, Office tables, laptop tables, ...
* File System
    /home -> /ritesh -> /Desktop
    /mnt -> /EFS, /hdd, /externaldrive -> 
    /share 
    /usr
```

```python
Terminology:-
* root -> top node of the tree # 1
* children -> left child and/or right child # 1 parent -> left child (7), right child (9)
* siblings -> 2 nodes which have the same parent # 2, 6 are siblings, common parent (7)
* leaf nodes -> these are nodes with 0 children # 2, 5, 11, and 5 leaf nodes
* internal nodes -> the ones which are *not* on the boundary of the tree # C and G
```

```python
Types of BT:-
1. Full Binary Tree - each node has either 0 or 2 children
2. Complete Binary Tree - each node has 2 children (leaf nodes is fine)
3. Balanced Binary Tree - each node has 2 children + its height balanced
        height balanced - difference between the left subtree and right subtree <= 1 (absolute)
        most optimized DS you can have
4. Skewed binary tree - neither of the 3 above # time consuming DS, not efficient

a b c d e
a -> b -> c -> d -> e
```

```python
def inorder_traversal(node): # LC -> P -> RC
    if not node:
        return
    
    inorder_traversal(node.left)
    print(node.data) # visiting
    inorder_traversal(node.right)
    
def preorder_traversal(node): # P -> LC -> RC
    if not node:
        return
    
    print(node.data) # visiting
    preorder_traversal(node.left)
    preorder_traversal(node.right)

def postorder_traversal(node): # LC -> RC -> P
    if not node:
        return
    
    postorder_traversal(node.left)
    postorder_traversal(node.right)
    print(node.data) # visiting
```

```python
Traversals:-
1. Level order traversal
    * Start from root node
    * for each level print the nodes left to right
    * TC = O(n) for n nodes in a BT
2. Inorder traversal # 2, 7, 6, 1, 9, 19
    * Start from root node
    * Print the left child -> parent -> right child
    * TC = O(n) for n nodes in a BT
3. Preorder traversal # 1, 7, 2, 6, 9, 19
    * Start from root node
    * Print the parent -> left child -> right child
    * TC = O(n) for n nodes in a BT
4. Postorder traversal # 2, 6, 7, 19, 9, 1
    * Start from root node
    * Print the left child -> right child -> parent
    * TC = O(n) for n nodes in a BT

print = visit (for all above)
```

```python
# insertion, deletion and search in a BT
class Node:
    def __init__(self, value):
        self.data = value
        self.left = None
        self.right = None
        
    def level_order_traversal(self): # 1, 7, 9, 2, 6, 19 - O(n)
        values = []
        values.append(self) # 6, 19
        
        while values:
            cur_node = values.pop(0) # 2
            print(cur_node.data, end=" -- ")
            
            if cur_node.left:
                values.append(cur_node.left)
            if cur_node.right:
                values.append(cur_node.right)
                
    # Cases
    # C1: No left child -> new node becomes the left child
    # C2: left child present but no right child -> new node becomes the right child
    def insert(self, value): # n nodes in your BT -> TC <= O(n) -> TC = O(n)
        new_node = Node(value)
        
        values = []
        values.append(self) # 9, 
        while values:
            cur_node = values.pop(0) # 7
            if cur_node.left:
                values.append(cur_node.left)
            else:
                cur_node.left = new_node
                return
            
            if cur_node.right:
                values.append(cur_node.right)
            else:
                cur_node.right = new_node
                return
    
    
#     def search(self, value): # TC = O(n)
    # do a LOT -> search for value in it.
    # if not found return -1
    
    
    # Principle
    # if you are deleting any node
    # locate the node you want to delete
    # locate the deepest right(if missing left) child node
    # make the deepest right(if missing left) child node as its substitute
    # Cases:-
    # C1 -> Delete root
    # C2 -> Delete parent
    # C3 -> Delete leaf node
    def delete(self, value):
        values = []
        values.append(self)
        
        # temp -> node that we want to delete
        # cur_node -> deepest node
        cur_node, temp = None, None
        while values:
            cur_node = values.pop(0)
            if cur_node.data == value: # root or parent
                temp = cur_node
                
            if cur_node.left:
                values.append(cur_node.left)
            if cur_node.right:
                values.append(cur_node.right)
            
        
        if not temp:
            print("Given value is not present in BT: %s" % (value))
            return
        
        val = cur_node.data
#         cur_node = None # fixed # get the parent of it and delete the child
        node_list = []
        node_list.append(self)
        while True:
            head = node_list.pop(0)
            if head.left is cur_node:
                head.left = None
                break
            else:
                node_list.append(head.left)
                
            if head.right is cur_node:
                head.right = None
                break
            else:
                node_list.append(head.right)
                
        temp.data = val
```
