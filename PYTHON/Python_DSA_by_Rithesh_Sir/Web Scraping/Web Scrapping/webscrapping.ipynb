```python
url = 'https://www.codewithharry.com/videos/'
page = r.get(url)
soup = bs(page.text, 'html.parser')
```

```python
c_name=[]
cn = soup.find_all('div', class_='title-font flex text-lg font-medium text-gray-900 mb-3')
for i in cn:
    c_name.append(i.text)
```

```python
c_name
```

```python
c_des=[]
cd= soup.find_all(class_='text-gray-700 text-base')
for i in cd:
    c_des.append(i.text)
```

```python
c_des
```

```python
info=[] 

for i in range(len(c_name)):
    info.append({'COURSE NAME': c_name[i], 'COURSE DESC': c_des[i]})
    
```

```python
info
```

```python
with open ('nirob.json', 'w') as f:  # [{},{},{},{}]
    json.dump(info, f)
    
```
