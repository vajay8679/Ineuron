```python
import bs4
import requests
```

```python
url ='https://www.w3schools.com/xml/xpath_examples.asp'
data = requests.get(url)
#print(data)
soup = bs4.BeautifulSoup(data.text,'html.parser')
#print(soup)
print(soup.prettify())
```

```python
for para in soup.find('a'):
    print(para)
```

```python
for para in soup.find_all('h2'):
    #print(para)
    print(para.text)
```

```python
url = 'https://en.wikipedia.org/wiki/Machine_learning'
res = requests.get(url)
print("Object",type(res))
soup1 = bs4.BeautifulSoup(res.text,'html.parser')
print("Object",type(soup1))
```

```python
for para1 in soup1.select('.mw-headline'):
    print(para1.text,end=" ")
```
